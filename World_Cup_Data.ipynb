```python
 import pandas as pd
```

```python
results = pd.read_csv('/content/results.csv', parse_dates=['date'])
```

```python
results
```

```python
results['date'].dt.year
```

```python
results_euro = results[results['tournament']== 'UEFA Euro qualification']
```

```python
results_euro['year'] = results_euro['date'].dt.year
```

```python
results_euro
```

```python
home_scores = results_euro[['home_team' , 'home_score' , 'year']].rename({'home_team' : 'team' , 'home_score' : 'score'}, axis= 'columns')
away_scores = results_euro[['away_team' , 'away_score' , 'year']].rename({'away_team' : 'team' , 'away_score' : 'score'}, axis= 'columns')
```

```python
home_scores
```

```python
away_scores
```

```python
scores_euro = pd.concat([home_scores , away_scores ]).reset_index(drop=True).fillna(0)
```

```python
scores_euro
```

```python
scores_euro.groupby(['team' , 'year']).score.sum().unstack().fillna(0)
```

```python
scores_euro_year = scores_euro.groupby(['team' , 'year']).score.sum().unstack().fillna(0).cumsum(axis=1).reset_index()
```

```python
scores_euro_year
```

```python
flags = pd.read_csv('/content/countries-fifa-flags.csv')
```

```python
flags
```

```python
scores_euro_year.merge(flags, left_on='team' , right_on='country' , how='left').to_csv('euro-data.csv' , index=False)
```
